# 1. 导入所需要的的第三方数据库

```python
import numpy as np
from sklearn.cluster import AffinityPropagation
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn import metrics
import pandas as pd
```

# 2. 导入sklearn库中的糖尿病病人数据

```python
diabetes = load_diabetes()
X = diabetes.data
```

# 3. 进行标准化处理，返回值为标准化后的数据

```python
from sklearn.preprocessing import StandardScaler

StandardScaler().fit_transform(X)
```

# 4. 进行区间缩放，返回值为缩放到[0, 1]区间的数据

```python
from sklearn.preprocessing import MinMaxScaler

MinMaxScaler().fit_transform(X)
```

# 5. 进行归一化处理，返回值为归一化后的数据

```python
from sklearn.preprocessing import Normalizer

Normalizer().fit_transform(X)
```

# 6. 通过迭代找出最佳的preference数值

```python
resq = []
for preference in np.arange(-50,0,1):
    ap = AffinityPropagation(random_state=0,preference = preference)
    ap.fit(X)
    n_clusters = len([i for i in set(ap.labels_)])
    stats = str(pd.Series([i for i in ap.labels_]).value_counts().values)
    resq.append({' preference': preference,'n_clusters':n_clusters,'stats':stats})
df = pd.DataFrame(resq)
df
```

# 7. 根据最佳preference，使用AffinityPropagation模型对数据进行拟合并将标签加入数据集中

```python
ap = AffinityPropagation(random_state=0,preference = -3)
ap.fit(X)
diabetes['ap_label'] = ap.labels_
```

# 8. 查看标签分类

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(ap.labels_)
le.classes_
```

# 9. 通过对不同特征进行作图看出特征之间的关系

```python
plt.figure(figsize=(10,10)) 
colors = [ 'g', 'r'] 
markers = ['o', 'D']
for i,l in enumerate(ap.labels_): 
     plt.plot(X[i][2],X[i][0],color=colors[l],marker=markers[l],ls='None') 
plt.xlabel("身体混乱指数")
plt.ylabel("年龄")
plt.show()
```
