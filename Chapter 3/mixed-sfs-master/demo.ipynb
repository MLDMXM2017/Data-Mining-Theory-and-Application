```python
train2 = pd.read_table('vehicle.dat',sep=' ',names=['COMPACTNESS','CIRCULARITY','DISTANCE CIRCULARITY','RADIUS RATIO','PR.AXIS ASPECT RATIO','MAX.LENGTH ASPECT RATIO' ,'SCATTER RATIO','ELONGATEDNESS','PR.AXIS RECTANGULARITY','MAX.LENGTH RECTANGULARITY','SCALED VARIANCE(MAJOR AXIS)','SCALED VARIANCE(MINOR AXIS)','SCALED RADIUS OF GYRATION','SKEWNESS ABOUT(MAJOR AXIS)','SKEWNESS ABOUT(MINOR AXIS)','KURTOSIS ABOUT(MINOR AXIS)','KURTOSIS ABOUT(MAJOR AXIS)','HOLLOWS RATIO','classes'])
```

```python
train2.head()
```

```python
classes = list(np.unique(train2.classes))
train2.classes.replace(classes, list(np.arange(len(classes))), inplace=True)
y = train2['classes']
```

```python
train2 = train2.drop(columns = ['classes'])
```

```python
featureWeights_2 = sfs.selectFeatures(inData=train2 , maxNumClusters = 3)
```

```python
from sklearn import preprocessing
from sklearn import neighbors
from sklearn import svm
from sklearn import model_selection
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
predictors1 = pd.DataFrame(train2,columns=['SCATTER RATIO',"SCALED VARIANCE(MINOR AXIS)","ELONGATEDNESS" ])
```

```python
predictors2 = pd.DataFrame(train2,columns=['DISTANCE CIRCULARITY',"PR.AXIS RECTANGULARITY","CIRCULARITY" ])
```

```python
predictors3 = pd.DataFrame(train2,columns=['SCALED VARIANCE(MAJOR AXIS)',"COMPACTNESS","MAX.LENGTH RECTANGULARITY" ])
```

```python
predictors4 = pd.DataFrame(train2,columns=['SCALED RADIUS OF GYRATION',"RADIUS RATIO","HOLLOWS RATIO" ])
```

```python
predictors5 = pd.DataFrame(train2,columns=['KURTOSIS ABOUT(MAJOR AXIS)',"SKEWNESS ABOUT(MINOR AXIS)","KURTOSIS ABOUT(MINOR AXIS)" ])
```

```python
predictors6 = pd.DataFrame(train2,columns=['SKEWNESS ABOUT(MAJOR AXIS)',"PR.AXIS ASPECT RATIO","MAX.LENGTH ASPECT RATIO" ])
```

```python
score = []
for X in [predictors1,predictors2,predictors3,predictors4,predictors5,predictors6]:
    X_train,X_test,y_train,y_test = model_selection.train_test_split(X, y, test_size = 0.25, random_state = 1234)
    nolinear_svc = svm.SVC(kernel='rbf')
    nolinear_svc.fit(X_train,y_train)
    pred_svc = nolinear_svc.predict(X_test)
    score += [metrics.accuracy_score(y_test,pred_svc)]
importent = [1,2,3,4,5,6]
plt.plot(importent, score)
    
plt.show()
```
